```python
 def sel_u_best(u, xk):
    lam = len(xk)
    n = len(xk[0])
    fx = np.zeros(lam)
    u_best = np.zeros((lam,n))
    next_best = [[] for i in range(u)]
    for i in range(lam):
        fx[i] = fun(xk[i])
    sort_fx = sorted(range(lam), key=lambda k: fx[k])
    for i in range(0,u):
        next_best[i] = xk[sort_fx[i]]
    next_best = np.array(next_best)
    return next_best
```
